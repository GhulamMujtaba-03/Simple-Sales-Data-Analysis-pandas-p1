```python
import pandas as pd
df = pd.read_csv("sales_project_dataset.csv")
```

```python
#  STEP 1 Data Understanding (EDA
# Q1 How many rows and columns are there in the dataset?
df.shape
# Q2 What are the names of all columns?
df.columns
# Q3 What is the data type of each column?
df.dtypes
# Q4 Are there any missing values in the dataset? If yes, in which columns?
df.isnull().sum()
```

```python
# STEP 2 Basic Analysis
# Q1 What is the total sales generated?
df["Sales"].agg ( total_sale=("sum"))
# Q2 What is the average sales value per order?
df["Sales"].agg( avg_sale = ("mean"))
# Q3 Which order had the highest sales value?
df["Sales"].agg( highest_sale = ("max"))
# Q4 Which order had the lowest sales value?
df["Sales"].agg( lowest_sale = ("min"))
```

```python
# STEP 3 City Analysis
# Q1 Calculate total sales for each city.
df.groupby("City").agg( total_sale_per_City=("Sales","sum"))
# Q2 Which three cities generated the highest sales?
top_cities_sales = df.groupby("City").agg(
    total_sales=("Sales", "sum")
).sort_values(by="total_sales", ascending=False).head(3)
top_cities_sales
# Q3 Which city has the highest number of orders?
df.groupby("City").size().sort_values(ascending=False).head(1)
```

```python
# STEP 4 Category Analysis
# Q1 Calculate total sales for each product category.
df.groupby("Category").agg( total_sale_per_cat=("Sales","sum"))
# Q2 Which category generated the highest sales?
df.groupby("Category").size().sort_values(ascending=False).head(2)
# Q3 Which category generated the lowest sales?
df.groupby("Category").size().sort_values(ascending=True).head(2)
```

```python
# STEP 5 Product Analysis
# Q1 Which five products have the highest total sales?
df.groupby("Product").size().sort_values(ascending=False).head(5)
# Q2 Which products generated the least sales?
df.groupby("Product").size().sort_values(ascending=True).head(5)
# Q3 Which product was sold in the highest quantity?
df.groupby("Product").agg(total_quantity=("Quantity","sum")).sort_values(by="total_quantity",ascending=False).head(10)
```

```python
# STEP 6 Customer Analysis
# Q1 Which customers generated the highest total sales?
df.groupby("Customer").agg( total_sale=("Sales","sum")).sort_values(by="total_sale",ascending=False).head(5)
# Q2 Which customer placed the most orders?
df.groupby("Customer").size().sort_values(ascending=False).head(5)
# Q3 What is the average sales per customer?
df.groupby("Customer").agg(avg_sale=("Sales","mean"))
```

```python
# STEP 7 Time Analysis
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
# Q1 Calculate total sales for each year.
yearly_sales = df.groupby('Year')['Sales'].sum()
yearly_sales
# Q2 Calculate total sales for each month?
monthly_sales = df.groupby('Month')['Sales'].sum()
monthly_sales
# Q3 Which month had the highest total sales?
top_month = monthly_sales.sort_values(ascending=False).head(1)
top_month
```
